## 2. Univariate Analysis on Categorical Features

```python
for feature in categorical_features:
    if feature == 'tenure' or feature == 'customerID':
        pass
    else:
        df[feature].value_counts().plot.bar()
        plt.title(feature)
        plt.show()
```

## 3. Univariate Analysis on Numerical Features

```python
for feature in numerical_features:
    if feature == 'SeniorCitizen':
        df[feature].value_counts().plot(kind ='bar')
        plt.title(feature)
        plt.show()
    else:
        df[feature].plot(kind = 'kde')
        plt.title(feature)
        plt.show()
```

## 4. Multivariate Analysis

### 4.1 Categorical Features

```python
for feature in categorical_features:
    if feature == 'tenure' or feature == 'customerID' or feature == 'Churn':
        pass
    else:
        sns.countplot(data = df , x = 'Churn', hue = feature)
        plt.title(feature)
        plt.show()
```

### 4.2 Numerical Features

```python
sns.kdeplot(df.MonthlyCharges[(df["Churn"] == 'No')], color = 'Blue', shade = True)
sns.kdeplot(df.MonthlyCharges[df['Churn'] == 'Yes'], color = 'Red', shade= True)
plt.title('Monthly Charges by Churn')
plt.show()
sns.kdeplot(df.TotalCharges[df['Churn'] == 'Yes'], color = 'Purple', shade = True)
sns.kdeplot(df.TotalCharges[df['Churn'] == 'No'], color = 'Green', shade = True)
plt.title('Total Charges by Churn')
plt.show()
sns.boxplot(x=df['tenure'],y=df['Churn'])
plt.show()
```

## 5. Data Preprocessing

```python
#Let's look at our features again
features = df.columns
features
```

```python
#Here customerID is not useful for training our model. Let's drop it.
df.drop(['customerID'], axis = 1, inplace = True)
```

```python
#Let's convert our categorical features using get dummies
df = pd.get_dummies(df, drop_first = True)
```

```python
#Let's check top 10 values
df.head(10)
```

```python
#Let's see correlation between features
plt.figure(figsize = (10,5))
sns.heatmap(df.corr())
```

```python
#Let's split our dataset.
#Here our target column is Churn_yes
y = df['Churn_Yes']
x = df.drop('Churn_Yes', axis = 1)
```

```python
#Let's check our x and y
y
```

```python
x
```

```python
#importing min max scalar for scaling down the feature
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
scaler = preprocessing.MinMaxScaler()
minmax_df = scaler.fit_transform(x)
x = pd.DataFrame(minmax_df, columns = x.columns)
```

```python
#Let's check our x
x
```

```python
#Let's split our dataset into training and testing
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y, test_size = 0.3, random_state = (0))
```

```python
#Let's check shape of our splitted dataset
print('X_train :', X_train.shape, 'X_test:', X_test.shape, 'y_train:', y_train.shape, 'y_test:',y_test.shape)
```

## 6. Building Model

```python
#Let's import different classification and regression models from sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

### 6.1 Support Vector Machine

```python
svm = SVC()
svm.fit(X_train,y_train)
accuracy = svm.score(X_test,y_test)
print('The accuracy of SVM is:', (accuracy * 100).round(1),'%')
```

```python
y_predict = svm.predict(X_test)
svmac = accuracy_score(y_test,y_predict)
print('The accuracy score is:', svmac)
cm = confusion_matrix(y_test,y_predict)
plt.figure(figsize = (10,10))
sns.heatmap(cm, annot = True, fmt =" ")
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('SVM Confusion Matrix')
```

### 6.2 Logistic Regression

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
accuracy = lr.score(X_test,y_test)
print('The accuracy of Logistic Regression Model is :', round((accuracy * 100),2), '%')
```

```python
y_predict = lr.predict(X_test)
lrac = accuracy_score(y_test,y_predict)
print('The accuracy score is:', lrac)
cm = confusion_matrix(y_test,y_predict)
plt.figure(figsize = (10,10))
sns.heatmap(cm, annot = True, fmt =" ")
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Logistic Regression Confusion Matrix')
```

### 6.3 KNeighbors Classifier

```python
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
accuracy = knn.score(X_test,y_test)
print('The accuracy of KNN is:',round((accuracy *100),2),'%')
```

```python
y_predict = knn.predict(X_test)
knnac =  accuracy_score(y_test,y_predict)
print('The accuracy score is:',knnac)
cm = confusion_matrix(y_test,y_predict)
plt.figure(figsize = (10,10))
sns.heatmap(cm, annot = True, fmt = " ")
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('KNN Confusion Matrix')
```

### 6.4 Decision Tree

```python
tree = DecisionTreeClassifier()
tree.fit(X_train,y_train)
accuracy = tree.score(X_test,y_test)
print('The accuracy of Desision Tree is:',round((accuracy *100),2),'%')
```

```python
y_predict = tree.predict(X_test)
tac = accuracy_score(y_test,y_predict)
print('The accuracy score is:',tac)
cm = confusion_matrix(y_test,y_predict)
plt.figure(figsize = (10,10))
sns.heatmap(cm, annot = True, fmt =" ")
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.title('Decision Tree Confusion Matrix')
```

### 6. 5 Random Forest

```python
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
accuracy = rf.score(X_test,y_test)
print('The accuracy of Random Forest is:',round((accuracy*100),2),'%')
```

```python
y_predict = rf.predict(X_test)
rfac = accuracy_score(y_test,y_predict)
print('The accuracy score is:',rfac)
cm = confusion_matrix(y_test,y_predict)
plt.figure(figsize = (10,10))
sns.heatmap(cm, annot = True, fmt = " ")
plt.xlabel('Preicted Values')
plt.ylabel('Actual Values')
plt.title('Ranom Forest Confusion Matrix')
```

### 7. Conclusion

Let's see which is best model.

```python
models = {'Support Vector Machine' : [svm.score(X_test,y_test)],
         'Logistic Regression': [lr.score(X_test,y_test)],
         'KNearest Neighbor': [knn.score(X_test,y_test)],
         'Decision Tree': [tree.score(X_test,y_test)],
         'Random Forest': [rf.score(X_test,y_test)]}
result = pd.DataFrame.from_dict(models,orient = 'index',columns = ['Score'])
result.style.highlight_max(color = 'Orange',axis = 0)
```
